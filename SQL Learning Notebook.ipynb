**3\. Create Table Script;**

this script is used for creating tables

````sql
CREATE TABLE EMPLOYE_DETAILS
(
    EID NUMERIC(4) PRIMARY KEY,
    ENAME VARCHAR(15) NOT NULL,
    JOB VARCHAR(15) NOT NULL,
    PHONE_NO NUMERIC(10) UNIQUE,
    AGE NUMERIC(2) NOT NULL CHECK(AGE>0),
    SAL NUMERIC(5) CHECK(SAL>0),
    MGR NUMERIC(4)
);
````

**Wildcard Commands:**

```
SELECT * FROM EMPLOYE_DETAILS where EID='2121'

```



```
SELECT * FROM EMPLOYE_DETAILS where JOB like '%T'

```



```
SELECT * FROM EMPLOYE_DETAILS where ENAME like '_ICKY'
```

**'\_' is used to represent a single character**



```
SELECT * FROM EMPLOYE_DETAILS where ENAME like '[vas]%'

```

**\- Command above returns all names STARTING with 'v' 'a' or 's' ... anything can follow b/c '%'**

**'\[\]' returns result if any characters inside get match**
